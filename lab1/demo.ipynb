```python
import numpy as np
from utils import generate_linear, generate_XOR_easy, show_result
from model import Model
```

```python
x, y = generate_linear(100)
# x, y = generate_XOR_easy()
```

```python
x[:80,:]
```

```python
model = Model(input_dim=2, hidden_units1=4, hidden_units2=4, output_dim=1)
```

```python
model.train(x, y, epochs=100, batchsize=10, learning_rate=0.1, loss_function='binary_cross_entropy')
model.plot_loss_curve()
```

```python
y_pred_list = []
loss_average = 0
print("Test Predictions:")
for xi, y_true in zip(x, y):
    y_pred, y_hat, loss = model.predict(xi.reshape(1, -1), y_true.reshape(1, -1))
    print("Input: {:.4f}, Predicted: {:.4f}, Ground Truth: {}".format(xi[0], y_hat[0][0], y_true[0]))
    y_pred_list.append(y_pred[0][0])
    loss_average += loss
accuracy = np.mean(y_pred_list == y.squeeze()) * 100
loss_average /= len(y)
print("Loss: {:.4f} Accuracy: {:.4f}%".format(loss_average, accuracy))
show_result(x, y, y_pred_list)
```
